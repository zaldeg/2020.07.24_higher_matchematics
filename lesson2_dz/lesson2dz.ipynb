Решите систему уравнений и неравенств:
- y = x2 – 1
- exp(x) + x∙(1 – y) > 1


```python
from scipy.optimize import fsolve
def f(x):
    return (np.exp(x) - x*x*x + 2*x - 1)
answers = set()
for i in range(-10, 10 , 1):
    answers.add(round(fsolve(f, i)[0], 5))
answers
x = list(answers)
y = np.array(x) * np.array(x)  - 1
for i in range(len(x)):
    print(f'корень {i + 1}: [{x[i]}, {y[i]}]')
```
